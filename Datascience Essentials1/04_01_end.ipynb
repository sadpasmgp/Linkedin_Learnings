```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

### Creating a line chart from a list object

#### Plotting a line chart in matplotlib

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

plt.plot(x,y)
```

```python
x1=np.sin(x)
y1=np.cos(y)
plt.plot(x1, y1)
plt.plot(x, y)
```

#### Plotting a line chart from a Pandas object

```python
address = r'C:\Users\sp\Downloads\09Python Datascience Essentials part1\Datascience Essentials1 Exercise Files\Data\mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg = cars['mpg']
```
