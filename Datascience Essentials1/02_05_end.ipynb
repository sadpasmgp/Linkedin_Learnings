### Grouping data by column index

```python
address = r'C:\Users\sp\Downloads\09Python Datascience Essentials part1\Datascience Essentials1 Exercise Files\Data\mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.head()
```

```python
cars_groups = cars.groupby(cars['cyl'])
cars_groups.mean()
```

```python
cars.groupby(cars['cyl'])
```

```python
cars.groupby(cars['cyl']).sum()
```

```python
cars.groupby(cars['am']).sum()
```

```python
cars.groupby(cars['am']).mean()
```
